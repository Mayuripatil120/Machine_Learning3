#### Step 2- Seprate X and y (weight)

#### Step 4 - Forward Feature selection 

#### categorical pipe - Ordinal Encoder

```python
cat = list(x.columns[x.dtypes== "object"])
con = list(x.columns[x.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = make_pipeline(
    SimpleImputer(strategy ="mean"),
    StandardScaler()
    
)
```

```python
cat_pipe1 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder()
)
```

```python
pre1 = ColumnTransformer(
    [
        ("num", num_pipe1,con),
        ("cat", cat_pipe1, cat)
    ]
).set_output(transform="pandas")
```

```python
x_pre =pre1.fit_transform(x)
x_pre.head()
```

#### Feature Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
model1 = LinearRegression()
sel = SequentialFeatureSelector(model1,n_features_to_select="auto", direction="backward")
sel.fit_transform(x_pre,y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols
```

```python
sel_cols[0]
```

```python
sel_cols[0].split("__")
```

```python
sel_cols[0].split("__")[1]
```

```python
imp_cols =[col.split("__")[1] for col in sel_cols]
imp_cols
```

```python
x_sel = x[imp_cols]
x_sel.head()
```

#### create a final pipeline for x

##### cat- OneHotEncoder



```python
cat_sel = list(x_sel.columns[x_sel.dtypes=="object"])
con_sel = list(x_sel.columns[x_sel.dtypes != "object"])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()

)
```

```python
cat_pipe2= make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre2 = ColumnTransformer(
    [
        ("num", num_pipe2 , con_sel),
        ("cat",cat_pipe2,cat_sel)
    ]
).set_output(transform="pandas")
```

```python
x_sel.head()
```

```python
x_sel_pre =pre2.fit_transform(x_sel)
x_sel_pre.head()
```

#### Step 5- train test spilt

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_sel_pre, y, test_size=0.2, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

#### Step 6 - Bulid Model

```python
final_model = LinearRegression()
final_model.fit(xtrain,ytrain)
```

```python
final_model.coef_
```

#### Step 7 - Evaluate Model

```python
final_model.score(xtrain,ytrain)
```

```python
final_model.score(xtest,ytest)
```

```python
from sklearn.metrics import(
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model,x,y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y,ypred)
    mae= mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2=r2_score(y,ypred)
    print(f"RMSE:{rmse:.2f}")
    print(f"MAE:{mae:.2f}")
    print(f"MAPE:{mape:.4f}")
    print(f"R2 score:{r2:.4f}")
```

```python
print("Train Result:")
evaluate_model(final_model,xtrain,ytrain)
```

```python
print("Test Results:")
evaluate_model(final_model,xtest,ytest)
```
